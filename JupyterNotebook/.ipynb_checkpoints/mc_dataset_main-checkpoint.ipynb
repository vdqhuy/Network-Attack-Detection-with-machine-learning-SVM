```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.offline as pyo
import tensorflow as tf
from tensorflow import keras
from keras import regularizers
import xgboost as xgb
from sklearn.decomposition import PCA
from sklearn import tree
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
%matplotlib inline
import os
import gc
```

```python
df_main = pd.DataFrame(data=None)

for dirname, _, filenames in os.walk('../Dataset/IDS 2018 Intrusion CSVs (CSE-CIC-IDS2018)/MC/'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        
        # Loại bỏ các tệp, thư mục cần bỏ qua
        if 'readme.txt' in file_path or 'MainDataset' in file_path:
            continue
        
        print(file_path)
        df_temp = pd.read_csv(file_path)
        df_main = pd.concat([df_main, df_temp], ignore_index=True)
        del df_temp
        gc.collect()
```

Tiền xử lý dữ liệu

```python
df_main['Label'] = df_main['Label'].apply(lambda x: 0 if 'benign' in x.lower() else
                                                    (2 if 'dos' in x.lower() else
                                                     (3 if 'brute' in x.lower() else
                                                      (4 if 'sql injection' in x.lower() else
                                                       (5 if 'bot' in x.lower() else
                                                        (6 if 'infilteration' in x.lower() else
                                                         1))))))

df_main['Label'].value_counts()
```

```python
df_main.isna().sum().sum()
```

```python
# Loại bỏ các giá trị trùng lặp và cột hằng (constant columns)
df_main.drop_duplicates(inplace=True)
variances = df_main.var(numeric_only=True)
constant_columns = variances[variances == 0].index
df_main.drop(constant_columns, axis=1, inplace=True)

# In ra các cột hằng và hình dạng của DataFrame sau khi xử lý
print(constant_columns)
print(df_main.shape)
```
